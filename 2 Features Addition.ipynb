#### get_zone_attraction_df<a class="anchor" id="get_zone_attraction_df"></a>

```python
# Change Transport to Transport (Sentosa Express) or Transport (Sentosa Beach Shuttle/Tram)
# Or combine Transport (Sentosa Express) and Transport (Sentosa Beach Shuttle/Tram) into Transport
def get_zone_attraction_df(df):
    
    """
    Create time of the day for the date column
    Inputs      : df      = attraction_df, list of all attractions in the data provided such as F&B, attraction, etc.
    outputs     : zone_df, updated data frame total attraction summarized and score normalized between 1-10
    """   
    
    zone_df=df.groupby(by='Zone')[['Lat','Long']].mean().reset_index()

    zone_attraction_df=df[df.Sector.isin(['Attraction','F&B','Hotel','Transport (Sentosa Express)'])].pivot_table(index=['Zone'],columns='Sector',values='Places/ Location',aggfunc='count',fill_value=0).reset_index()
    zone_attraction_df['Total_attraction']=zone_attraction_df['Attraction']*0.4+zone_attraction_df['F&B']*0.25+zone_attraction_df['Hotel']*0.25+zone_attraction_df['Transport (Sentosa Express)']*0.10

    zone_df=pd.merge(zone_df,zone_attraction_df, on='Zone', how='left')
    zone_df['Total_attraction']=zone_df['Total_attraction'].fillna(0)
    
    zone_df['Total_attraction_Normalized']=(zone_df['Total_attraction']-zone_df['Total_attraction'].min())/(zone_df['Total_attraction'].max()-zone_df['Total_attraction'].min())*(10-1)+1
    
    return zone_df
```

#### get_bus_zone_distance<a class="anchor" id="get_bus_zone_distance"></a>

```python
def get_bus_zone_distance(df1,df2):
    
    """
    Calculate distance in metres between two lat and long positions
    Inputs      : df1      = bus_df,containing list of unique bus stops and lat and long
                : df2      = zone_df, the data frame created above with list of all zones and their lat and long
    outputs     : updated df1 with the closest zone to the bus stop marked
    """  
    
    R = 6371
    
    for i in range(len(df2)):

        dlat=np.radians(df1.Lat - df2.Lat[i])
        dlong=np.radians(df1.Long - df2.Long[i])

        a = np.sin(dlat / 2)**2 + np.cos(np.radians(df2.Lat[i])) * np.cos(np.radians(df1.Lat)) * np.sin(dlong / 2)**2
        c = 2 * np.arctan2(np.sqrt(a), np.sqrt(1 - a))
        distance = R * c    
        df1[df2.Zone[i]]=np.round(distance,3)*1000
        
    df1['Zone'] = df1[df2.Zone.unique()].idxmin(axis=1)
    df1=pd.merge(df1,df2[['Zone','Attraction','F&B','Total_attraction','Total_attraction_Normalized']], on='Zone', how='left')

    return df1
```

#### get_total_guests<a class="anchor" id="get_total_guests"></a>

```python
def get_total_guests(df):
    
    df['Total_Guests']=df['Total In']+df['Total Out']

    return df
```

#### get_attractiveness_factor_busstop<a class="anchor" id="get_attractiveness_factor_busstop"></a>

```python
def get_attractiveness_factor_busstop(df):
    
    """
    get popularity factor for each bus stops based on the closeby attractions to the bus stop and guests demand at the stop
    Inputs      : df      = bus-df with list of stops and merged with total guests demand in the total data provided
    outputs     : updated data frame with a new column of "Popularity_factor"
    """   
    
    
    df['Total_Guests_Normalized']=((df.Total_Guests-df.Total_Guests.min())/(df.Total_Guests.max()-df.Total_Guests.min()))*(10-1)+1
    
    df['Popularity_factor']=0.6*df['Total_Guests_Normalized']+0.4*df['Total_attraction_Normalized']
    
    df['Popularity_factor']=((df.Popularity_factor-df.Popularity_factor.min())/(df.Popularity_factor.max()-df.Popularity_factor.min()))*(10-1)+1
    
    return df
```

#### get_route_df<a class="anchor" id="get_route_df"></a>

```python
def get_route_df(df):
    route_df=df[['Route','Bus Stops','Route Sequence']].copy()
    route_df['next_stop']=np.where((route_df['Route Sequence'].shift(-1)<route_df['Route Sequence'])|
                               route_df['Route Sequence'].shift(-1).isnull(),
                               'Beach Station',
                               route_df['Bus Stops'].shift(-1))
    return route_df
```

#### get_network_nodes_edges<a class="anchor" id="get_network_nodes_edges"></a>

```python
def get_network_nodes_edges(bus_df,sdc_df):
    
    """
    Create unique nodes and edges from the bus data frame; nodes represent all bus stops and edges represent # guests transferring between two bus stops
    Inputs      : bus_df      = Unique list of all bus stops
                : sdc_df      = sentosa ridership data frame
    outputs     : Nodes and edges data frame
    """

    nodes_df=(bus_df[['Bus Stops','Popularity_factor']].drop_duplicates()).copy()

    edge_df=sdc_df[sdc_df['Total_Guests']>=1].groupby(by=['Bus Stop','next_stop'])['Total_Guests'].sum().reset_index()
    edge_df=edge_df.rename(columns={'Bus Stop':'from','next_stop':'to','Total_Guests':'weight'})
    edge_df=edge_df[~(edge_df['from']==edge_df['to'])]
    
    edge_df['weight']=(edge_df.weight-edge_df.weight.min())/(edge_df.weight.max()-edge_df.weight.min())*10
    

    return nodes_df, edge_df
```

#### get_pos_dict<a class="anchor" id="get_pos_dict"></a>

```python
def get_pos_dict(df):
    
    """
    Get a dictionary for each node i.e., the bus stop and their lat and long. The dictionary will be used to plot the bus stops on the network map
    Inputs      : df          = a dataframe with list of unique bus stops and their lat and long columns                
    outputs     : Dictionary with nodes i.e., bus stop names and their lat/long
    """
    
    routes_lats=df[['Bus Stops','Long','Lat']]
    xy=routes_lats.drop_duplicates().copy()
    xy['pos']=list(zip(xy.Long,xy.Lat))
    routes_lats=xy[['Bus Stops','pos']]
    routes_lats=routes_lats.set_index('Bus Stops')
    routes_lats=routes_lats.to_dict('index')
    pos_dict={}


    for key,value in routes_lats.items():
        for key2,value2 in value.items():
            pos_dict[key] = np.asarray(value2)
            
    return pos_dict
```

#### get_busstops_guests_hourly<a class="anchor" id="get_busstops_guests_hourly"></a>

```python
def get_busstops_guests_hourly(df):    
    
    """
    Create a dataframe which summarizes the total guests visiting each bus stop in a given route every hour
    Inputs      : df          = sdc_df i.e., bus ridership cleaned data
    outputs     : df_hourly, new data frame with hourly level total guests at a given bus stop in a given route 
    """
    

    df_hourly=df[( (df.TIME_OF_DAY.isin(['Morning','Afternoon','Evening'])))
                    ].groupby([pd.Grouper(key='Open Timestamp', freq='H'), 
                                pd.Grouper(key='Route'),
                                   pd.Grouper(key='Bus Stop'),
                                      pd.Grouper(key='WEEKEND'),])[['Total_Guests','Total In','Total Out']].sum().reset_index()

    df_hourly['HOUR_OF_DAY']=df_hourly['Open Timestamp'].dt.hour
    
    return df_hourly
```

#### get_pub_sch_hols

```python
def get_pub_sch_hols(df):    
    """
    Create a dataframe which adds public and school holidays
    Inputs      : df          = sdc_df i.e., bus ridership cleaned data
    outputs     : df with public and school holidays flag added
    """
    df['Is_Pub_Hol']=np.where((df['Open Timestamp'].dt.date).isin(pub_hols['Date'].dt.date),1,0)
    df['Is_Sch_Hol']=np.where((df['Open Timestamp'].dt.date).isin(sch_hols['Date'].dt.date),1,0)
    
    return df
```

```python
def get_event_flag(df,events,bus_stop=True):
    """
    Create a dataframe which adds events
    Inputs      : df          = sdc_df i.e., bus ridership cleaned data
    outputs     : df with event flag added
    """
    if bus_stop:
        df = df.merge(events, 
                      left_on=['YEAR','MONTH_OF_YEAR','DAY_OF_MONTH','HOUR_OF_DAY','Bus Stop'], 
                      right_on=['YEAR','MONTH_OF_YEAR','DAY_OF_MONTH','HOUR_OF_DAY','Bus Stop'],
                      how='left').set_index(df.index)
        # fill NaN values in Is_Event column with 0
        df.fillna(value={'Is_Event':0},inplace=True)
        # convert Is_Event column from float to int
    #    df['Is_Event'] = df['Is_Event'].astype(int)
    else:
        df = df.merge(events, 
                      left_on=['YEAR','MONTH_OF_YEAR','DAY_OF_MONTH','HOUR_OF_DAY'], 
                      right_on=['YEAR','MONTH_OF_YEAR','DAY_OF_MONTH','HOUR_OF_DAY'],
                      how='left').set_index(df.index)
        # fill NaN values in Is_Event column with 0
        df.fillna(value={'Is_Event':0},inplace=True)
        # convert Is_Event column from float to int
    #    df['Is_Event'] = df['Is_Event'].astype(int)
    
    return df
```

```python
def get_time_difference(df):
    """
    Adds a time difference feature, measuring the time between stops for buses
    Inputs      : df = ridership dataframe 
    outputs     : model data frame with added feature
    """
    
    df=df.sort_values(by=['YEAR','MONTH_OF_YEAR','DAY_OF_MONTH','Bus No','Open Timestamp']).copy()
    
    df['Time_Difference'] = df['Open Timestamp']-df['Open Timestamp'].shift()
    df['Time_Difference'] = df.apply (lambda row: pd.Timedelta(row['Time_Difference']).seconds, axis=1)
    
    # Set rows where the buses restarts to 0 time difference
    df['Time_Difference'] = np.where(df['Bus No']!=df['Bus No'].shift(),0,df['Time_Difference'])
    df['Time_Difference'] = np.where((df['Time_Difference']>900),0,df['Time_Difference'])

    df=df.sort_index()
    
    return df
```

```python
def get_pax_on_bus(df):
    """
    Adds a pax on bus feature, measuring number of people on the bus
    Inputs      : df = ridership dataframe 
    outputs     : model data frame with added feature
    """
    
    df=df.sort_values(by=['YEAR','MONTH_OF_YEAR','DAY_OF_MONTH','Bus No','Open Timestamp']).copy()
    df['Pax_On_Bus'] = 0
    df['Pax_On_Bus'] = df['Pax_On_Bus'].shift()+df['Total In']-df['Total Out']
    df['Pax_On_Bus'] = np.where(df['Bus No']!=df['Bus No'].shift(),0,df['Pax_On_Bus'])

    df['Pax_On_Bus'] = np.where(df['Pax_On_Bus'].shift()+df['Total In']-df['Total Out']<0,0,df['Pax_On_Bus'].shift()+df['Total In']-df['Total Out'])
#    df['Pax_On_Bus'] = df.apply (lambda row: np.where(row['Pax_On_Bus'].shift()+row['Total In']-row['Total Out']<0,0,row['Pax_On_Bus'].shift()+row['Total In']-row['Total Out']))
#     df['Pax_On_Bus'] = df['Pax_On_Bus'].shift()+df['Total In']-df['Total Out']
#     df['Pax_On_Bus'] = np.where(df['Bus No']<0,0,df['Pax_On_Bus'])   
#     df['Pax_On_Bus'] = df['Pax_On_Bus'].shift()+df['Total In']-df['Total Out']
   # df['Pax_On_Bus'] = np.where(df['Bus No']<0,0,df['Pax_On_Bus'])   
#     df['Pax_On_Bus'] = np.where(df['Time_Difference']-df['Time_Difference'].shift()>900,0,df['Time_Difference'])
    
    df['Pax_On_Bus']=df['Pax_On_Bus'].replace(np.nan,0)
#     df=df.sort_index()
    
    return df
```

```python
def get_trip_number(rdf,bdf):
    """
    Adds a feature that assigns a new number for each trip a particular bus takes. 
    A trip is defined by every time a bus arrives back at Route sequence 1
    Inputs      : rdf = ridership dataframe,  bdf = bus stops dataframe
    outputs     : model data frame with added feature
    """
    rdf=rdf.merge(bdf[['Route','Bus Stops','Route Sequence']],
                  right_on=['Route','Bus Stops'],
                  left_on=['Route','Bus Stop'],
                  how='left')
    rdf=rdf.sort_values(by=['YEAR','MONTH_OF_YEAR','DAY_OF_MONTH','Bus No','Open Timestamp'])
    
    rdf['Trip'] = ((rdf['Route Sequence'] == 1) & (rdf['Route Sequence'].shift() != 1.0))
    rdf['Trip'] = rdf['Trip'].shift()
    rdf['Trip'] = np.where((rdf['Bus No'] != rdf['Bus No'].shift()),True,rdf['Trip'])
    rdf['Trip'] = np.where(((rdf['Route Sequence']==1)&(rdf['Time_Difference']>900)),True,rdf['Trip'])    

    rdf['Trip'].replace(np.nan,False,inplace=True)
    rdf['Trip']=rdf['Trip'].cumsum()
    rdf['Trip']= rdf['Trip'].replace(True,1)
    rdf=rdf.sort_index()
    
    return rdf
```

```python
def combine_multiple_door_openings(df,bdf):
    """
    Converts ridership dataframe with multiple door openings to a single entry taking the time from the first entry
    Inputs      : df = ridership dataframe,  bdf = bus stops dataframe
    outputs     : converted ridership dataframe
    """
    
    df=df.sort_values(by=['YEAR','MONTH_OF_YEAR','DAY_OF_MONTH','Bus No','Open Timestamp'])
    df['first_door']=(df['Bus Stop']!=df['Bus Stop'].shift()) |((df['Route Sequence']==1)&(df['Time_Difference']==0))

    df['first_door'].replace(np.nan,False,inplace=True)
    df['first_door'] = df['first_door'].cumsum()
    df['first_door'] = df['first_door'].replace(True,1)
    
    sum_features=['Adult In', 'Adult Out','Child In', 'Child Out','Total In', 'Total Out','Total_Guests','Time_Difference']
    cat_features=df.drop(['first_door']+sum_features,axis=1).columns.unique()
    ap={tuple(sum_features):'sum',tuple(cat_features):'first',}
    agg_params = {x:v for k, v in ap.items() for x in k}

    rdf_new_grp=df.groupby('first_door').agg(agg_params)
    rdf_new_grp=rdf_new_grp.reset_index()
    rdf_new_grp=rdf_new_grp.drop('first_door',axis=1)
    rdf_new_grp=get_time_difference(rdf_new_grp)
    rdf_new_grp=rdf_new_grp.drop(['Route Sequence','Bus Stops'],axis=1)
    rdf_new_grp=get_trip_number(rdf_new_grp,bdf)
    
    return rdf_new_grp
```
